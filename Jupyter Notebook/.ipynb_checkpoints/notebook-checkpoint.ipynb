```python
import pandas as pd 
import gspread
import seaborn
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import textwrap
import numpy as np
from itertools import chain
from oauth2client.service_account import ServiceAccountCredentials
```

```python
control = pd.read_csv("../Responses/Control Group.csv")
experimental = pd.read_csv("../Responses/Experimental Group.csv")
```

```python
def convertLikert(df):
    likertMap = {"Strongly Disagree": 1, "Disagree": 2, "Neutral": 3, "Agree": 4, "Strongly Agree": 5}
    df[likertColumns] = df[likertColumns].replace(likertMap)
    
likertColumns = list(control[control.columns[control.isin(["Strongly Disagree","Disagree","Neutral","Agree","Strongly Agree"]).all()]].columns)

convertLikert(control)
convertLikert(experimental)
```

```python
datetimeColumns = ["Timestamp"]

control[datetimeColumns] = pd.to_datetime(control[datetimeColumns].stack(),infer_datetime_format=True).unstack()
experimental[datetimeColumns] = pd.to_datetime(experimental[datetimeColumns].stack(),infer_datetime_format=True).unstack()

timeColumns = ["T1 - In Vscode", "T1 - Outside Vscode", "T1 - Reading Tutorial", "T1 - Web Browser", "T1 - Total Time","T2 - In Vscode", "T2 - Outside Vscode", "T2 - Reading Tutorial", "T2 - Web Browser", "T2 - Total Time","T3 - In Vscode", "T3 - Outside Vscode", "T3 - Reading Tutorial", "T3 - Web Browser", "T3 - Total Time"]

control[timeColumns] = control[timeColumns].stack().map(lambda value: pd.Timedelta("00:"+value).seconds).unstack()
experimental[timeColumns] = experimental[timeColumns].stack().map(lambda value: pd.Timedelta("00:"+value).seconds).unstack()

numberColumns = ["Age","... worked on that had a Dockerfile?", "... used Dockerfiles created by others (colleagues or third parties)?", "... created/updated a Dockerfile?", "I have a good enough understanding of English so that I can confidently answer this survey and understand sentences with Docker and Unix related terms.", "T1 - It was simple to understand the objective of the task and the task's instructions.", "T2 - It was simple to understand the objective of the task and the task's instructions.", "T3 - It was simple to understand the objective of the task and the task's instructions." ,"T1 - In Vscode (Switches)", "T1 - Outside Vscode (Switches)", "T1 - Reading Tutorial (Switches)", "T1 - Web Browser (Switches)","T2 - In Vscode (Switches)", "T2 - Outside Vscode (Switches)", "T2 - Reading Tutorial (Switches)", "T2 - Web Browser (Switches)","T3 - In Vscode (Switches)", "T3 - Outside Vscode (Switches)", "T3 - Reading Tutorial (Switches)", "T3 - Web Browser (Switches)"]

control[numberColumns] = pd.to_numeric(control[numberColumns].stack()).unstack()
experimental[numberColumns] = pd.to_numeric(experimental[numberColumns].stack()).unstack()

commaSeparatedColumns = ["In the Dockerfiles I've developed, I've specified...","T1 - Which of these features did you use to solve this task? (If you don't remember the name of a specific feature, the instructions document contains a list with all the features using the same names.)","T2 - Which of these features did you use to solve this task? (If you don't remember the name of a specific feature, the instructions document contains a list with all the features using the same names.)","T3 - Which of these features did you use to solve this task? (If you don't remember the name of a specific feature, the instructions document contains a list with all the features using the same names.)"]

control[commaSeparatedColumns] = control[commaSeparatedColumns].stack().map(lambda value: value.split(",")).unstack()
experimental[commaSeparatedColumns] = experimental[commaSeparatedColumns].stack().map(lambda value: value.split(",")).unstack()
```

```python
control["group"] = "control"
experimental["group"] = "experimental"
both = pd.concat([control, experimental])
both = both.reset_index()
both["id"] = both.index

both = both.add_prefix('value')
both.rename(columns={'valueid': 'id'}, inplace=True)
both.rename(columns={'valuegroup': 'group'}, inplace=True)

both = pd.wide_to_long(both,stubnames="value",i=["id"],j="params",suffix='.+')
both.reset_index(inplace=True)
both = both.explode('value')

dataset = both
```

```python
timeColumns.extend(["T1 - Valuable Time", "T2 - Valuable Time", "T3 - Valuable Time"])

t1ValuableTimes = []
t2ValuableTimes = []
t3ValuableTimes = []
for id in dataset[dataset["params"] == "T1 - In Vscode"].id:
    group = dataset[(dataset["params"] == "T1 - In Vscode") & (dataset["id"] == id)]["group"].values[0]
    t1ValuableTime = int(dataset[(dataset["params"] == "T1 - In Vscode") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T1 - Outside Vscode") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T1 - Web Browser") & (dataset["id"] == id)]["value"])
    t2ValuableTime = int(dataset[(dataset["params"] == "T2 - In Vscode") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T2 - Outside Vscode") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T2 - Web Browser") & (dataset["id"] == id)]["value"])
    t3ValuableTime = int(dataset[(dataset["params"] == "T3 - In Vscode") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T3 - Outside Vscode") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T3 - Web Browser") & (dataset["id"] == id)]["value"])
    t1ValuableTimes.append({"id": id, "params": "T1 - Valuable Time", "group": str(group), "value": t1ValuableTime})
    t2ValuableTimes.append({"id": id, "params": "T2 - Valuable Time", "group": str(group), "value": t2ValuableTime})
    t3ValuableTimes.append({"id": id, "params": "T3 - Valuable Time", "group": str(group), "value": t3ValuableTime})

dataset = dataset.append(t1ValuableTimes)
dataset = dataset.append(t2ValuableTimes)
dataset = dataset.append(t3ValuableTimes)
dataset.reset_index(drop=True, inplace=True)
```

```python
numberColumns.extend(["T1 - Total Switches", "T2 - Total Switches", "T3 - Total Switches", "T1 - Total Switches Per Minute", "T2 - Total Switches Per Minute", "T3 - Total Switches Per Minute"])

t1TotalSwitches = []
t2TotalSwitches = []
t3TotalSwitches = []
t1TotalSwitchesPerMinute = []
t2TotalSwitchesPerMinute = []
t3TotalSwitchesPerMinute = []
for id in dataset[dataset["params"] == "T1 - In Vscode"].id:
    group = dataset[(dataset["params"] == "T1 - In Vscode") & (dataset["id"] == id)]["group"].values[0]
    t1TotalSwitch = int(dataset[(dataset["params"] == "T1 - In Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T1 - Outside Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T1 - Reading Tutorial (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T1 - Web Browser (Switches)") & (dataset["id"] == id)]["value"])
    t2TotalSwitch = int(dataset[(dataset["params"] == "T2 - In Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T2 - Outside Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T2 - Reading Tutorial (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T2 - Web Browser (Switches)") & (dataset["id"] == id)]["value"])
    t3TotalSwitch = int(dataset[(dataset["params"] == "T3 - In Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T3 - Outside Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T3 - Reading Tutorial (Switches)") & (dataset["id"] == id)]["value"]) + \
                    int(dataset[(dataset["params"] == "T3 - Web Browser (Switches)") & (dataset["id"] == id)]["value"])
    t1TotalSwitchPerMinute = (t1TotalSwitch / int(dataset[(dataset["params"] == "T1 - Total Time") & (dataset["id"] == id)]["value"])) * 60
    t2TotalSwitchPerMinute = (t2TotalSwitch / int(dataset[(dataset["params"] == "T2 - Total Time") & (dataset["id"] == id)]["value"])) * 60
    t3TotalSwitchPerMinute = (t3TotalSwitch / int(dataset[(dataset["params"] == "T3 - Total Time") & (dataset["id"] == id)]["value"])) * 60
    t1TotalSwitches.append({"id": id, "params": "T1 - Total Switches", "group": str(group), "value": t1TotalSwitch})
    t1TotalSwitches.append({"id": id, "params": "T2 - Total Switches", "group": str(group), "value": t2TotalSwitch})
    t1TotalSwitches.append({"id": id, "params": "T3 - Total Switches", "group": str(group), "value": t3TotalSwitch})
    t1TotalSwitchesPerMinute.append({"id": id, "params": "T1 - Total Switches Per Minute", "group": str(group), "value": t1TotalSwitchPerMinute})
    t1TotalSwitchesPerMinute.append({"id": id, "params": "T2 - Total Switches Per Minute", "group": str(group), "value": t2TotalSwitchPerMinute})
    t1TotalSwitchesPerMinute.append({"id": id, "params": "T3 - Total Switches Per Minute", "group": str(group), "value": t3TotalSwitchPerMinute})
    
dataset = dataset.append(t1TotalSwitches)
dataset = dataset.append(t2TotalSwitches)
dataset = dataset.append(t3TotalSwitches)
dataset = dataset.append(t1TotalSwitchesPerMinute)
dataset = dataset.append(t2TotalSwitchesPerMinute)
dataset = dataset.append(t3TotalSwitchesPerMinute)
dataset.reset_index(drop=True, inplace=True)
```

```python
numberColumns.extend(["In Vscode - Total Switches", "Outside Vscode - Total Switches", "Reading Tutorial - Total Switches", "Web Browser - Total Switches"])

vscodeTotalSwitches = []
outsideTotalSwitches = []
tutorialTotalSwitches = []
browserTotalSwitches = []
for id in dataset[dataset["params"] == "T1 - In Vscode"].id:
    group = dataset[(dataset["params"] == "T1 - In Vscode") & (dataset["id"] == id)]["group"].values[0]
    vscodeTotalSwitch = int(dataset[(dataset["params"] == "T1 - In Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                        int(dataset[(dataset["params"] == "T2 - In Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                        int(dataset[(dataset["params"] == "T3 - In Vscode (Switches)") & (dataset["id"] == id)]["value"])
    outsideTotalSwitch = int(dataset[(dataset["params"] == "T1 - Outside Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                         int(dataset[(dataset["params"] == "T2 - Outside Vscode (Switches)") & (dataset["id"] == id)]["value"]) + \
                         int(dataset[(dataset["params"] == "T3 - Outside Vscode (Switches)") & (dataset["id"] == id)]["value"])
    tutorialTotalSwitch = int(dataset[(dataset["params"] == "T1 - Reading Tutorial (Switches)") & (dataset["id"] == id)]["value"]) + \
                          int(dataset[(dataset["params"] == "T2 - Reading Tutorial (Switches)") & (dataset["id"] == id)]["value"]) + \
                          int(dataset[(dataset["params"] == "T3 - Reading Tutorial (Switches)") & (dataset["id"] == id)]["value"])
    browserTotalSwitch = int(dataset[(dataset["params"] == "T1 - Web Browser (Switches)") & (dataset["id"] == id)]["value"]) + \
                         int(dataset[(dataset["params"] == "T2 - Web Browser (Switches)") & (dataset["id"] == id)]["value"]) + \
                         int(dataset[(dataset["params"] == "T3 - Web Browser (Switches)") & (dataset["id"] == id)]["value"])
    vscodeTotalSwitches.append({"id": id, "params": "In Vscode - Total Switches", "group": str(group), "value": vscodeTotalSwitch})
    outsideTotalSwitches.append({"id": id, "params": "Outside Vscode - Total Switches", "group": str(group), "value": outsideTotalSwitch})
    tutorialTotalSwitches.append({"id": id, "params": "Reading Tutorial - Total Switches", "group": str(group), "value": tutorialTotalSwitch})
    browserTotalSwitches.append({"id": id, "params": "Web Browser - Total Switches", "group": str(group), "value": browserTotalSwitch})

dataset = dataset.append(vscodeTotalSwitches)
dataset = dataset.append(outsideTotalSwitches)
dataset = dataset.append(tutorialTotalSwitches)
dataset = dataset.append(browserTotalSwitches)
dataset.reset_index(drop=True, inplace=True)
    
```

```python
#https://stackoverflow.com/a/51535326/6391820
def show_values_on_bars(axs):
    def _show_on_single_plot(ax):        
        for p in ax.patches:
            value = '{:.0f}'.format(p.get_width())
            _x = p.get_x() + p.get_width() - 0.6
            _y = p.get_y() + p.get_height() - 0.30
            ax.text(_x, _y, value, ha="center", color="white") 

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)


data = dataset

group = "control"

param1 = "T1 - Which of these features did you use to solve this task? (If you don't remember the name of a specific feature, the instructions document contains a list with all the features using the same names.)"
param2 = "T2 - Which of these features did you use to solve this task? (If you don't remember the name of a specific feature, the instructions document contains a list with all the features using the same names.)"
param3 = "T3 - Which of these features did you use to solve this task? (If you don't remember the name of a specific feature, the instructions document contains a list with all the features using the same names.)"

mask = (data['group'] == group) & ((data['params'] == param1) | (data['params'] == param2) | (data['params'] == param3))

plt.figure(figsize=(7,2.8))
occurences = pd.DataFrame(data[mask]['value'].str.strip().value_counts())
occurences.reset_index(inplace=True)
occurences.rename(columns={'index':'responses', 'value': 'count'}, inplace=True)

occurences.loc[occurences["responses"]=="Terminal",'count'] = int(occurences[occurences["responses"]=="Terminal"]['count']) + int(occurences[occurences["responses"]=="terminal"]['count']) + int(occurences[occurences["responses"]=="CLI"]['count'])
occurences = occurences[(occurences['responses'] != "terminal") & (occurences['responses'] != "CLI")]
occurences.loc[occurences["responses"]=="Web search",'count'] = int(occurences[occurences["responses"]=="Web search"]['count']) + int(occurences[occurences["responses"]=="internet"]['count'])
occurences = occurences[(occurences['responses'] != "internet")]
occurences = occurences.sort_values(by=['count'], ascending=False)
ax = seaborn.barplot(x="count", y="responses", data=occurences, color="C0")
ax.set(yticklabels=[textwrap.fill(e, 24) for e in occurences['responses'].unique()],ylabel='', xlabel='Responses', xlim=(0, 31))
show_values_on_bars(ax)
plt.show()
```

```python
from scipy.stats import ttest_ind, mannwhitneyu, shapiro, levene
import numpy as np

def runTtest(paramToCompare, axis=0, equal_var=False, nan_policy='propagate'):
    ctrl = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "control")]
    expr = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "experimental")]

    return ttest_ind(ctrl['value'], expr['value'], axis, equal_var, nan_policy) 

def runWmw(paramToCompare, use_continuity=True, alternative=None):
    ctrl = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "control")]
    expr = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "experimental")]
    
    return mannwhitneyu(ctrl['value'], expr['value'], use_continuity, alternative)

def runShapiro(paramToCompare):
    data = dataset[dataset['params'] == paramToCompare]
    
    return shapiro(data['value'])

def runLevene(paramToCompare, center='median', proportiontocut=0.05):
    ctrl = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "control")]
    expr = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "experimental")]
    
    return levene(ctrl['value'], expr['value'], center=center, proportiontocut=proportiontocut)

def runTOST(paramToCompare, bound=2):
    ctrl = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "control")]
    expr = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "experimental")]
    
    df_ttesta = mannwhitneyu(expr['value'] + bound, ctrl['value'], alternative='greater')
    df_ttestb = mannwhitneyu(ctrl['value'] + bound, expr['value'], alternative='greater')
    return max(df_ttestb.pvalue, df_ttesta.pvalue)

def runMean(paramToCompare):
    ctrl = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "control")]
    expr = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "experimental")]
    
    return np.mean(ctrl['value']),np.mean(expr['value'])
    
def runStd(paramToCompare):
    ctrl = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "control")]
    expr = dataset[(dataset['params'] == paramToCompare) & (dataset["group"] == "experimental")]

    return np.std(ctrl['value']),np.std(expr['value'])

ctrl_mean = []
ctrl_std = []
expr_mean = []
expr_std = []
u = []
twosided_pvalues = []
less_pvalues = []
greater_pvalues = []
for param in numberColumns + timeColumns + likertColumns:
    mean = runMean(param)
    std = runStd(param)
    twosided_res = runWmw(param, alternative="two-sided")
    less_res = runWmw(param, alternative="less")
    greater_res = runWmw(param, alternative="greater")
    twosided_pvalues.append(format(twosided_res.pvalue,'f'))
    less_pvalues.append(format(less_res.pvalue,'f'))
    greater_pvalues.append(format(greater_res.pvalue,'f'))
    u.append(twosided_res.statistic)
    ctrl_mean.append(mean[0])
    ctrl_std.append(std[0])
    expr_mean.append(mean[1])
    expr_std.append(std[1])

ttestResults = pd.DataFrame ({
    'params': numberColumns + timeColumns + likertColumns,
    "ctrl_mean": ctrl_mean,
    "ctrl_std": ctrl_std,
    "expr_mean": expr_mean,
    "expr_std": expr_std,
    "u" : u, 
    "twosided_pvalues" : twosided_pvalues, 
    "less_pvalues" : less_pvalues,
    "greater_pvalues" : greater_pvalues 
})
ttestResults
```
